```python
import pandas as pd

# Example input
df = pd.DataFrame({
    'user_id': ['A', 'A', 'A','A','B','B','A'],
    'book_code': ['B1', 'B2', 'B3','B4','B2','B3','B1'],
    'category_name': [['G1', 'G2'], ['G2', 'G3'], ['G4'],['G2','G4'],['G2', 'G3'], ['G4'],['G1', 'G2']],
    'book_create_dt': pd.to_datetime(['2025-07-05', '2025-07-08', '2025-07-10','2025-07-12','2025-07-08', '2025-07-10','2025-07-9']),
    'book_type': ['RTM', 'RTE', 'RTM','RTM', 'RTE', 'RTM','RTM'],
    'language': ['EN', 'FR', 'EN','EN', 'FR', 'EN','EN'],
   
})

# Ensure category_name is a list
df['category_name'] = df['category_name'].apply(lambda x: x if isinstance(x, list) else x.split(','))

# Unique values
all_categories = sorted(set(cat for sublist in df['category_name'] for cat in sublist))
all_book_types = sorted(df['book_type'].unique())
all_languages = sorted(df['language'].unique())

# Sort
df = df.sort_values(by=['user_id', 'book_create_dt']).reset_index(drop=True)

# Output list
output = []

# Process per user
for user_id, user_df in df.groupby('user_id', sort=False):
    last_read_cat = {cat: None for cat in all_categories}
    read_count_cat = {cat: 0 for cat in all_categories}
    
    last_read_type = {btype: None for btype in all_book_types}
    read_count_type = {btype: 0 for btype in all_book_types}

    last_read_lang = {lang: None for lang in all_languages}
    read_count_lang = {lang: 0 for lang in all_languages}

    total_books_read = 0  # Track number of books read so far by user

    for _, row in user_df.iterrows():
        book_dt = row['book_create_dt']
        book_type = row['book_type']
        language = row['language']
        current_categories = row['category_name']

        row_out = {
            'user_id': user_id,
            'book_code': row['book_code'],
            'book_create_dt': book_dt,
        }

        # Categories
        for cat in all_categories:
            count = read_count_cat[cat]
            row_out[f'{cat}_count'] = count
            row_out[f'{cat}_days'] = (book_dt - last_read_cat[cat]).days if last_read_cat[cat] else 0
            row_out[f'{cat}_pct'] = (count / total_books_read) if total_books_read > 0 else 0

        # Book types
        for btype in all_book_types:
            count = read_count_type[btype]
            row_out[f'{btype}_count'] = count
            row_out[f'{btype}_days'] = (book_dt - last_read_type[btype]).days if last_read_type[btype] else 0
            row_out[f'{btype}_pct'] = (count / total_books_read) if total_books_read > 0 else 0

        # Languages
        for lang in all_languages:
            count = read_count_lang[lang]
            row_out[f'{lang}_count'] = count
            row_out[f'{lang}_days'] = (book_dt - last_read_lang[lang]).days if last_read_lang[lang] else 0
            row_out[f'{lang}_pct'] = (count / total_books_read) if total_books_read > 0 else 0

        # Update state
        for cat in current_categories:
            read_count_cat[cat] += 1
            last_read_cat[cat] = book_dt

        read_count_type[book_type] += 1
        last_read_type[book_type] = book_dt

        read_count_lang[language] += 1
        last_read_lang[language] = book_dt

        total_books_read += 1

        output.append(row_out)

# Final DataFrame
result_df = pd.DataFrame(output)
```

```python
import pandas as pd
import torch
import torch.nn as nn

def encode_column_with_vocab(series: pd.Series, col_name: str, existing_vocab=None):
    """
    Encodes a single categorical column for embedding.
    Returns encoded indices, vocab, inverse map.
    """
    series = series.fillna('unk').astype(str)

    if existing_vocab is None:
        unique_vals = sorted(series.unique().tolist())
        if 'unk' not in unique_vals:
            unique_vals.append('unk')
        vocab_map = {val: idx for idx, val in enumerate(unique_vals)}
    else:
        vocab_map = existing_vocab

    inverse_map = {idx: val for val, idx in vocab_map.items()}
    encoded_series = series.map(lambda x: vocab_map.get(x, vocab_map['unk']))
    encoded_series.name = f"{col_name}_idx"
    return encoded_series, vocab_map, inverse_map


def create_embedding_layer(vocab_size, embedding_dim):
    """
    Creates a PyTorch embedding layer with given vocab size and dimension.
    """
    return nn.Embedding(num_embeddings=vocab_size, embedding_dim=embedding_dim)


def get_embeddings_for_column(encoded_series, embedding_layer):
    """
    Returns embedding vectors for each value in the encoded column.
    """
    input_tensor = torch.tensor(encoded_series.tolist(), dtype=torch.long)
    return embedding_layer(input_tensor)
```

```python
# Sample data
df = pd.DataFrame({
    'grade': ['A', 'B', None, 'C', 'A']
})

# Step 1: Encode the column
encoded_col, grade_vocab, grade_inv_vocab = encode_column_with_vocab(df['grade'], 'grade')
df['grade_idx'] = encoded_col

# Step 2: Create embedding layer
embedding_dim = 4  # can be 4, 8, 16 etc.
embedding_layer = create_embedding_layer(vocab_size=len(grade_vocab), embedding_dim=embedding_dim)

# Step 3: Get embeddings for the encoded column
embeddings = get_embeddings_for_column(df['grade_idx'], embedding_layer)

print("Embedding matrix shape:", embeddings.shape)  # [num_rows, embedding_dim]
print("Embeddings:\n", embeddings)
```

```python
import pandas as pd
from datetime import datetime

def add_days_since_column(df: pd.DataFrame, date_col: str, output_col: str = None, current_date=None):
    """
    Adds a new column representing number of days between current date and a given date column.

    Parameters:
    - df: input DataFrame
    - date_col: name of the column with datetime values
    - output_col: name of the new column (optional)
    - current_date: custom "now" date (optional, defaults to today)

    Returns:
    - DataFrame with new column added
    """
    if output_col is None:
        output_col = f'days_since_{date_col}'

    if current_date is None:
        current_date = pd.Timestamp.now()

    df[date_col] = pd.to_datetime(df[date_col], errors='coerce')
    df[output_col] = (current_date - df[date_col]).dt.days

    return df

add_days_since_column(user_class,'teacher_create_dt','teacher_days-from_signup')
```

```python
import pandas as pd

def one_hot_encode_column(df: pd.DataFrame, column: str, prefix: str = None, drop_original: bool = False):
    """
    One-hot encodes a single column in the DataFrame.

    Parameters:
    - df: Input DataFrame
    - column: Name of the column to encode
    - prefix: Prefix for new columns (defaults to column name)
    - drop_original: If True, drops the original column

    Returns:
    - df with one-hot encoded columns added
    """
    if prefix is None:
        prefix = column

    # Fill nulls with 'unk'
    df[column] = df[column].fillna('unk').astype(str)

    # Create one-hot encoded DataFrame
    dummies = pd.get_dummies(df[column], prefix=prefix)

    # Concatenate back
    df = pd.concat([df, dummies], axis=1)

    if drop_original:
        df = df.drop(columns=[column])

    return df


one_hot_encode_column(user_class,'class_activation_bucket')
```
